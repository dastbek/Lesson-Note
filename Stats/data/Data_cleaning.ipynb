```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("world/gini.csv")
```

```python
df
```

```python
year_list = np.arange(2020-5*10,2025,5)
```

```python
year_list = year_list.tolist()
```

```python
for i in range(len(year_list)):
    year_list[i] = str(year_list[i])
```

```python
year_list.insert(0, 'country')
```

```python
year_list
```

```python
df = df.loc[:, year_list]
```

```python
df
```

```python
df.melt(id_vars=["country"], 
        var_name="Year", 
        value_name="Gini")
```
